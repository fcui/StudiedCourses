```python
x = np.arange(0, 2, 0.01)
plt.figure()
plt.plot(x, 0 * x, x, 0 * x + 1)
# plt.plot(x, 0 * x + 1)
plt.plot(x, x)
x = np.arange(1, 2, 0.01)
plt.plot(x, x - 1)
plt.xlabel('$z$')
plt.ylabel('$x$')
plt.title('$x$的取值范围')
```

### (2)

```python
x = np.arange(0, 1.1, 0.01)
plt.figure()
plt.plot(x, 0 * x, x, 0 * x + 1)
plt.plot(x, x)
plt.plot(0 * x, x)
plt.xlabel('$z$')
plt.ylabel('$x$')
plt.title('$x$的取值范围')
```
