### (2)

```r
F <- function(x) {
    val <- numeric(length(x))
    val[x < 0] <- 0
    val[x >= 2] <- 1
    val[(x >= 0) & (x < 1)] <- (x^2/2)[(x >= 0) & (x < 1)]
    val[(x >= 1) & (x < 2)] <- (-x^2/2 + 2 * x - 1)[(x >= 1) & (x < 2)]
    val
}
```

```r
x = seq(-1, 3, 0.01)
y = F(x)
library("ggplot2")
qplot(x, y, geom = 'line')
```
